```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(["A", "B", "C", "D"]
G.add_edges_from([("A", "B"), ("A", "C"), ("B", "D")])
nx.draw(G, with_labels=True, node_color="lightblue", node_size=2000, font_size=14, edge_color="gray")
plt.show()
```

```python
DG = nx.DiGraph()
DG.add_edges_from([("A", "B"), ("B", "C"), ("C", "A")])

nx.draw(DG, with_labels=True, node_color="orange", node_size=2000, font_size=14, arrows=True)
plt.show()
```

```python
WG = nx.Graph()
WG.add_edge("A", "B", weight=4)
WG.add_edge("A", "C", weight=2)
WG.add_edge("B", "C", weight=5)

pos = nx.spring_layout(WG)
nx.draw(WG, pos, with_labels=True, node_color="lightgreen", node_size=2000, font_size=14)

labels = nx.get_edge_attributes(WG, 'weight')
nx.draw_networkx_edge_labels(WG, pos, edge_labels=labels)

plt.show()
```

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_vertex(self, v):
        if v not in self.graph:
            self.graph[v] = []

    def add_edge(self, u, v):
        self.add_vertex(u)
        self.add_vertex(v)
        self.graph[u].append(v)
        self.graph[v].append(u)

    def display(self):
        for v, neighbors in self.graph.items():
            print(f"{v}: {neighbors}")
```

```python
def visualize_graph(graph):
    plt.figure(figsize=(6,6))

    # Assign positions (circle layout)
    nodes = list(graph.graph.keys())
    n = len(nodes)
    positions = {}
    for i, node in enumerate(nodes):
        angle = 2 * 3.14159 * i / n
        positions[node] = (np.cos(angle), np.sin(angle))

    # Draw edges
    for u in graph.graph:
        for v in graph.graph[u]:
            if u < v:  # to avoid double drawing
                x_values = [positions[u][0], positions[v][0]]
                y_values = [positions[u][1], positions[v][1]]
                plt.plot(x_values, y_values, "k-")

    # Draw nodes
    for node, (x, y) in positions.items():
        plt.scatter(x, y, s=1000, c="lightblue", edgecolors="black", zorder=2)
        plt.text(x, y, node, fontsize=12, ha="center", va="center", zorder=3)

    plt.axis("off")
    plt.show()
```

```python
import numpy as np

g = Graph()
g.add_edge("A", "B")
g.add_edge("A", "C")
g.add_edge("B", "D")
g.add_edge("C", "D")

g.display()
visualize_graph(g)
```
